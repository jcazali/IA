```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
img = cv2.imread('moedas.jpg', cv2.IMREAD_COLOR) 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
gray_blurred = cv2.blur(gray, (3, 3)) 
detected_circles = cv2.HoughCircles(gray_blurred,  
                   cv2.HOUGH_GRADIENT, 2, 20, param1 = 160, 
               param2 = 300, minRadius = 0, maxRadius = 0) 
```

```python
print(detected_circles)
```

```python
circles = np.uint16(np.around(detected_circles))
for i in circles[0,:]:
    cv2.circle(img, (i[0], i[1]), i[2], (0,255,0), 5)
cv2_imshow(img)
```
